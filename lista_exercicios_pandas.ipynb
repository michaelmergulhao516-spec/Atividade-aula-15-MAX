## 2. Seleção e Filtragem

### Questão 3

### Questão 4

## 3. Estatística Descritiva

### Questão 1

Calcule média, mediana, mínimo, máximo e desvio padrão.

```python
print("Média:")
print(df.mean(numeric_only=True))

print("\nMediana:")
print(df.median(numeric_only=True))

print("\nMínimo:")
print(df.min(numeric_only=True))

print("\nMáximo:")
print(df.max(numeric_only=True))

print("\nDesvio padrão:")
print(df.std(numeric_only=True))
```

### Questão 2

Identifique os registros acima da média.

```python
acima_media = df[df["radius_worst"] > df["radius_worst"].mean()]

print(acima_media)
```

### Questão 3

Identifique os registros abaixo da mediana.

```python
abaixo_mediana = df[
    df["texture_worst"] < df["texture_worst"].median()
]

print(abaixo_mediana)
```

### Questão 4

Qual coluna numérica possui maior variabilidade?

```python
variabilidade = df.std(numeric_only=True)

coluna_maior = variabilidade.idxmax()
valor_maior = variabilidade.max()

print(f"Coluna com maior variabilidade: {coluna_maior}")
print(f"Desvio padrão: {valor_maior}")
```
